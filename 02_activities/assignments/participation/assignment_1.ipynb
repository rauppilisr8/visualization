```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)  # seed for reproducibility
x = np.arange(50)
y = np.random.randint(0, 100,50)

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y) # Scatterplot
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y) # Vertical bar graph
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.barh(x,y) # Horizontal bar graph
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y) # Histogram (counts number of occurances)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y) # line chart
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
# Defining font specs in a python dictionary
font1 = {'family':'sans-serif' ,'color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time' , fontdict = font1 , loc = 'left') # added font and title alignment/ location
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, marker='*' , color = 'indigo') # Customizing marker.
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*' , color = 'indigo' ,linestyle = '--', linewidth = 2) # Customizing marker and line width & style.
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*' ,
markersize = 12, 
color = '#7425b9',
linestyle = '--',
linewidth = 2,
markeredgecolor = '#fa9359', 
markerfacecolor = '#000000' ) # Using Hex code for color instead & adding edge and fill colors
ax.grid() #adding grid lines
```

```python
# adding grid line size and color

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*' ,
markersize = 12, 
color = '#7425b9',
linestyle = '--',
linewidth = 2,
markeredgecolor = '#fa9359', 
markerfacecolor = '#000000' ) # Using Hex code for color instead & adding edge and fill colors

ax.grid(color = "skyblue" , linewidth = 1, linestyle = '-.')
```
